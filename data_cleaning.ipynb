# Data Cleaning

```python
#import csv file
data=pd.read_csv('C:/Users/Anne_2/Documents/Ironhack/Fifa_Moneyball-Case-Study/fifa21_male2.csv', dtype={78: 'str'})
```

```python
#view data file
pd.set_option('max_columns', None)
data.head()
```

```python
print(data.columns.tolist())
```

```python
#keeping only columns relevant for analysis and a few descriptives
data = data.filter(['ID', 'Name', 'Age', 'OVA', 'Nationality', 'Club', 'BP', 'Height', 'Weight', 'foot', 'Value', 'Wage', 'Crossing', 'Finishing', 'Heading Accuracy', 'Short Passing', 'Volleys', 'Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control', 'Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance', 'Shot Power', 'Jumping', 'Stamina', 'Strength', 'Long Shots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking', 'Standing Tackle', 'Sliding Tackle', 'GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Gender'], axis=1)
```

```python
#standardizing column names
data.columns=[i.lower().replace(' ','_') for i in data.columns]
```

```python
#Checking number of entries and for duplicates on ID
print(data.shape)
print('Duplicate ids:', data['id'].duplicated().sum())
```

```python
#checking for null values
data.isna().sum()
```

```python
data['position'] = str
data['position'] = np.where(data['bp'] == 'LS', 'forward', data['bp'])
```

```python
#Especially high number of missing values in 'composure'.
#Building categories of player positions and then check if null values concerning only certain positions
#if not, deleting all data rows with null values
data['bp'].value_counts()
```

```python
#categories: forward, midfielder, defender, keeper
#new variable position, derived from column bp (best position)
def recode_position(x):
    if (x == 'ST' or x == 'CF'):
        return 'forward'
    elif (x == 'CB' or x == 'LB' or x =='RB' or x == 'RWB' or x == 'LWB'):
        return 'defender'
    elif x == 'GK':
        return 'keeper'
    else:
        return 'midfielder'

data['position'] = data['bp'].apply(recode_position)
```

```python
#check recoding
pd.crosstab(data['bp'], data['position'])
```

```python
#check missing values within groups
data.groupby(['position'])[['id','composure']].count()
```

```python
#missing values seem to be randomly spread across groups - deleting all rows with missing values
data=data.dropna(axis=0)
data.shape
```

```python
#checking datatypes of columns
data.dtypes
```

```python
#checking on column 'value', it should be numerical for analysis
data['value'].head(5)
```

```python
#delete '€' sign and split column into number and unit
data['value']=[i.replace('€','') for i in data['value']]
data['unit'] = data['value'].str[-1:]
data['number']=[i.replace('K','') for i in data['value']]
data['number']=[i.replace('M','') for i in data['number']]
data['number']=data['number'].astype(float)
```

```python
data['number'].head()
```

```python
data['unit'].value_counts()
```

```python
#multiply number column by unit
data['unit']=[i.replace('0', '1') for i in data['unit']]
data['unit']=[i.replace('K', '1000') for i in data['unit']]
data['unit']=[i.replace('M', '1000000') for i in data['unit']]
data['unit']=data['unit'].astype(int)

data['market_value'] = (data['unit']*data['number']).astype(int)
```

```python
data['market_value'].head()
```

```python
data = data.drop(['unit', 'number'], axis=1)
```

```python
#dividing market value again by 1.000.000 to make number more readable
data['market_value']=data['market_value']/1000000
data['market_value'].head()
```

```python
#checking values of numeric variables, especially all skills should be between 0 and 100 (score)
data.describe()
```

####  Market value seems to have some extreme outliers as well as entries with a market value of '0'. It is not clear, if a market value of '0' is a missing value or an actual 'not worth anything value' or players, that are not available to the market anymore for whatever reasons... For analysis of skills and their impact on market value, reducing the dataset to players that: A) have a market value at all, B) don't belong to extreme outliers, since these extreme market values most likely are produced by factors other than basic playing skills

```python
# market value before dropping
data['market_value'].describe()
```

```python
#skipping data with market value = 0
data=data[data['market_value']>0]
data['market_value'].describe()
```

```python
#removing outliers of market_value
iqr = np.percentile(data['market_value'], 75) - np.percentile(data['market_value'],25)
lower_limit= np.percentile(data['market_value'],25)-1.5*iqr
upper_limit= np.percentile(data['market_value'], 75)+1.5*iqr
data=data[(data['market_value']>lower_limit)&(data['market_value']<upper_limit)]
data.reset_index(drop=True, inplace=True)
data['market_value'].describe()
```

```python
#saving cleaned data set
data.to_csv('C:/Users/Anne_2/Documents/Ironhack/Fifa_Moneyball-Case-Study/fifa21_cleaned.csv', index=False)
```
