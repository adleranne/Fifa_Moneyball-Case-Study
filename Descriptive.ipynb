# Descriptive statistics

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('C:/Users/Anne_2/Documents/Ironhack/Fifa_Moneyball-Case-Study/fifa21_cleaned.csv')
```

```python
# Number of players for each position
print(data.position.value_counts(normalize = True))
plt.figure(figsize=(5,2))
sns.histplot(data.position)
plt.show()
```

```python
# Market value for all players
sns.histplot(data['market_value'], bins=100)
plt.show()
```

```python
# Market value by position
data.groupby(['position'])['market_value'].mean().plot(kind='bar')
plt.show()
```
